```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout,GRU
from keras import optimizers
# Para la construcción de modelos usaremos esta biblioteca.
from keras.layers import LSTM


seed = 1234
np.random.seed(seed)
plt.style.use('ggplot')
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Para los diagramas usaremos esta biblioteca.
from sklearn.metrics import mean_squared_error, explained_variance_score
import matplotlib.pyplot as plt
from itertools import cycle
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
from pandas import read_csv
from matplotlib import pyplot
```

```python
dataraw = pd.read_csv('BTC-USD8y.csv',index_col='Date', parse_dates=['Date'])
dataraw
```

```python
dataraw.describe()
```

Selección de características

```python
# usar la función 'Date' y 'Close'

dataset = pd.DataFrame(dataraw[['Close']])
print(' Cantidad fila de datos: ',len(dataset))
#del dataset['Date']
fig = plt.figure(figsize=(14, 6))
plt.plot(dataset)
plt.xlabel('Fecha')
plt.ylabel('Precio del Bitcoin')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y")) # -%m-%d"))
plt.title('Precio de cierre de Bitcoin ')
plt.show()
```

Procesamiento de datos
Mormalizacion Min-Max

```python
#Min-Max Normalization
dataset_norm = dataset.copy()
dataset[['Close']]
scaler = MinMaxScaler()
dataset_norm['Close'] = scaler.fit_transform(dataset[['Close']])
dataset_norm
```

Normalizaciónn de datos - grafico

```python
fig = plt.figure(figsize=(14,6))
plt.plot(dataset_norm)
plt.xlabel('Date')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
plt.title('Normalización de los datos [0-1]')
plt.show()
```

Partición de los Datos

```python
# Partition data into data train, val & test
totaldata = dataset.values
totaldatatrain = int(len(totaldata)*0.7)
totaldataval = int(len(totaldata)*0.1)
totaldatatest = int(len(totaldata)*0.2)

# Store data into each partition
training_set = dataset_norm[0:totaldatatrain]
val_set=dataset_norm[totaldatatrain:totaldatatrain+totaldataval]
test_set = dataset_norm[totaldatatrain+totaldataval:]
```

Grafico de entrenamiento de los datos

```python
# graph of data training
fig = plt.figure(figsize=(14,6))
plt.plot(training_set)
plt.xlabel('Fecha')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
plt.title('Data de entrenamiento')
plt.show()
```

Grafico de validacion de los datos

```python
# graph of data validation
fig = plt.figure(figsize=(14,6))
plt.plot(val_set)
plt.xlabel('Fecha')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
plt.title('Data de validación')
val_set
```

Test de datos - grafico

```python
# graph of data test
fig = plt.figure(figsize=(10, 4))
plt.plot(test_set)
plt.xlabel('Fecha')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
plt.title('Data Prueba')
plt.show()
test_set
```

```python
# Create graph data test and prediction result
plt.figure(num=None, figsize=(10, 4), dpi=80,facecolor='w', edgecolor='k')
plt.title('Gráfico de la división de los datos')
plt.plot(training_set, color='green',label='Data de entrenamiento')
plt.plot(test_set , color='red',label='Data de Prueba')
plt.plot(val_set, color='blue',label='Data de Validación ')
plt.xlabel('Day')
#plt.ylabel('Price')
plt.legend()
plt.show()
```

Sliding Windows **Preparacion de los datos para la Red neuronal**

```python
# valor de inicialización de lag
# convertir una matriz de valores en una matriz de conjunto de datos
lag = 2
# Reduccir la complejidad
# sliding windows function
def create_sliding_windows(data,len_data,lag):
    x=[]
    y=[]
    for i in range(lag,len_data):
        x.append(data[i-lag:i,0])
        y.append(data[i,0])
    return np.array(x),np.array(y)

# Formateo de datos en una matriz para crear ventanas deslizantes
array_training_set = np.array(training_set)
array_val_set = np.array(val_set)
array_test_set = np.array(test_set)
print("set de arreglo de entrenamiento 70% ",array_training_set.shape)
print("set de arreglo de validación 10%",array_val_set.shape)
print("set de arreglo de prueba 20%",array_test_set.shape)
```

# remodelar la entrada para que sea [muestras, pasos de tiempo, características] que se requiere para LSTM


```python
# Create sliding windows into training data
x_train, y_train = create_sliding_windows(array_training_set,len(array_training_set), lag)
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
# Create sliding windows into validation data
x_val,y_val = create_sliding_windows(array_val_set,len(array_val_set),lag)
x_val = np.reshape(x_val, (x_val.shape[0],x_val.shape[1],1))
# Create sliding windows into test data
x_test,y_test = create_sliding_windows(array_test_set,len(array_test_set),lag)
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],1))
print("x_train",x_train.shape)
print("x_val",x_val.shape)
print("x_test",x_test.shape)
x_train
```

LSTM

```python
# Model
#batch size	neuronas	epocas
#  72	          100	    50
dropoutIndex = 0.0
neuronsLSTMLayer = 100
neuronsOutputLayer = 1
lossFunction = 'mean_squared_error'
#Optimizer = 'adam'
Epochs = 100
batchSize = 256 #32
Verbose = 1
learning_rate = 0.0001
model = Sequential()
model.add(LSTM(neuronsLSTMLayer, input_shape = (x_train.shape[1], x_train.shape[2])))
model.add(Dropout(dropoutIndex))
model.add(Dense(neuronsOutputLayer))
#model.compile(loss = lossFunction, optimizers.Adam(lr=learning_rate))

# Compiling the Gated Recurrent Unit
model.compile(optimizer= optimizers.Adam(lr=learning_rate),loss='mean_squared_error')

pred = model.fit(x_train, y_train, epochs = Epochs, batch_size = batchSize, validation_data = (x_test, y_test), verbose = Verbose, shuffle=False)
```

Modelo LSTM


Graph Training loss & Validation Loss

```python
# Graph model loss (train loss & val loss)
fig = plt.figure(figsize=(8, 4))
plt.plot(pred.history['loss'], label='pérdidas de entrenamiento')
plt.plot(pred.history['val_loss'], label='pérdidas de validación')
plt.title('Perdidas')
plt.ylabel('Perdidas')
plt.xlabel('Épocas')
plt.legend(loc='upper right')
plt.show()
```

Tabla de entrenamiento perdidas y perdidas de validation

```python
# Tabel value of training loss & validation loss
learningrate_parameter = learning_rate
train_loss=pred.history['loss'][-1]
validation_loss=pred.history['val_loss'][-1]
learningrate_parameter=pd.DataFrame(data=[[learningrate_parameter, train_loss, validation_loss]],
                                    columns=['Learning Rate', 'Training Loss', 'Validation Loss'])
learningrate_parameter.set_index('Learning Rate')
```

Modelo de implementación en prueba de datos

```python
# Implementation model into data test
y_pred_test = model.predict(x_test)

# Invert normalization min-max
y_pred_invert_norm = scaler.inverse_transform(y_pred_test)
y_pred_test

```

Comparacion de los datos de prueba con los resultados de prediccion

```python
# Comparison data test with data prediction
datacompare = pd.DataFrame()
datatest=np.array(dataset['Close'][totaldatatrain+totaldataval+lag:])
datapred= y_pred_invert_norm

datacompare['Data Test'] = datatest
datacompare['Prediction Results'] = datapred
datacompare
```

Prediction Results Evaluation

```python
# Calculatre value of Root Mean Square Error
def rmse(datatest, datapred):
    return np.round(np.sqrt(np.mean((datapred - datatest) ** 2)), 4)
print('Result Root Mean Square Error Prediction Model :',rmse(datatest, datapred))

def mape(datatest, datapred):
    return np.round(np.mean(np.abs((datatest - datapred) / datatest) ), 4)

print('Result Mean Absolute Percentage Error Prediction Model : ', mape(datatest, datapred), '%')
```

```python
# Create graph data test and prediction result
plt.figure(num=None, figsize=(10, 4), dpi=80,facecolor='w', edgecolor='k')
plt.title('Comparación de gráficos Datos reales y predicción de datos')
plt.plot(datacompare['Data Test'], color='red',label='Datos de entrenamiento')
plt.plot(datacompare['Prediction Results'], color='blue',label='Resultados de la predicción')
plt.xlabel('Dias')
plt.ylabel('Precio')
plt.legend()
plt.show()
```

```python
# make predictions
trainPredict = model.predict(x_train)
testPredict = model.predict(x_test)
# invert predictions
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([y_train])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([y_test])
# calculate root mean squared error
trainScore = np.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = np.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))
```
